**Ans3** - The difference in np.array(), np.asarray(), and np.asanyarray():

np.array(): This function creates a new array by copying the input data. It always returns a new array object, regardless of the type of input.
np.asarray(): This function converts the input into an array. If the input is already an array, it returns the same object without making a copy, making it more efficient when you want to avoid unnecessary copies.

np.asanyarray(): Similar to np.asarray(), this function converts the input to an array, but it preserves the input's subclass (like matrix or masked array) if applicable, while np.asarray() converts everything into a base ndarray.

The difference between deep copy and shallow copy:
Shallow Copy: In a shallow copy, a new array is created, but it still references the elements of the original array. Changes to the elements of the shallow copy will reflect in the original array.
Deep Copy: In a deep copy, a new array is created, and all elements are copied. Changes to the elements of the deep copy do not affect the original array, and vice versa.

```python
#Ans13
import matplotlib.pyplot as plt

x = np.random.rand(100)
y = np.random.rand(100)

plt.scatter(x, y, color='red', marker='o')

plt.axhline(y=0.5, color='blue', linestyle='--', label='y = 0.5')

plt.axvline(x=0.5, color='green', linestyle=':', label='x = 0.5')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.title('Advanced Scatter Plot of Random Values')

plt.legend()
plt.show()
```

```python
#Ans14
dates = pd.date_range(start='2023-01-01', periods=30, freq='D')
temperature = np.random.randint(20, 35, size=len(dates))
humidity = np.random.randint(40, 70, size=len(dates))

data = {'Date': dates, 'Temperature': temperature, 'Humidity': humidity}
df = pd.DataFrame(data)

fig, ax1 = plt.subplots(figsize=(10, 6))

ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature (°C)', color='tab:red')
ax1.plot(df['Date'], df['Temperature'], color='tab:red', label='Temperature')
ax1.tick_params(axis='y', labelcolor='tab:red')
plt.xticks(rotation=90)

ax2 = ax1.twinx()
ax2.set_ylabel('Humidity (%)', color='tab:blue')
ax2.plot(df['Date'], df['Humidity'], color='tab:blue', label='Humidity')
ax2.tick_params(axis='y', labelcolor='tab:blue')

plt.title('Temperature and Humidity Over Time')

fig.tight_layout()

plt.show()
```

```python
#Ans15&16
normal_data = np.random.randn(1000)

plt.hist(normal_data, bins=30, density=True, alpha=0.6, color='g')

mu, std = np.mean(normal_data), np.std(normal_data)
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = np.exp(-((x - mu) ** 2) / (2 * std ** 2)) / (std * np.sqrt(2 * np.pi))

plt.plot(x, p, 'k', linewidth=2)

plt.xlabel('Value')
plt.ylabel('Frequency/Probability')

plt.title('Histogram with PDF Overlay')
plt.show()
```

```python
#Ans17
import seaborn as sns

x_scatter = np.random.randn(100)
y_scatter = np.random.randn(100)

quadrants = ['Quadrant 1' if x > 0 and y > 0 else
             'Quadrant 2' if x < 0 and y > 0 else
             'Quadrant 3' if x < 0 and y < 0 else
             'Quadrant 4'
             for x, y in zip(x_scatter, y_scatter)]

sns.scatterplot(x=x_scatter, y=y_scatter, hue=quadrants, palette='Set1')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Quadrant-wise Scatter Plot')
plt.legend(title='Quadrants')
plt.show()
```

```python
#Ans18
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

x = np.linspace(0, 4 * np.pi, 100)
y = np.sin(x)

p = figure(title="Sine Wave Function", x_axis_label='X', y_axis_label='Y')

p.line(x, y, legend_label="Sine Wave", line_width=2)

p.xgrid.grid_line_color = 'gray'
p.ygrid.grid_line_color = 'gray'

output_notebook()
show(p)
```

```python
#Ans19
from bokeh.transform import factor_cmap
from bokeh.models import HoverTool

categories = ['A', 'B', 'C', 'D', 'E']
values = np.random.randint(10, 100, size=len(categories))

p = figure(x_range=categories, title="Random Categorical Bar Chart",
           x_axis_label='Category', y_axis_label='Value')

p.vbar(x=categories, top=values, width=0.8, color=factor_cmap('x', palette='Viridis256', factors=categories))

hover = HoverTool()
hover.tooltips = [("Category", "@x"), ("Value", "@top")]
p.add_tools(hover)

output_notebook()
show(p)
```
